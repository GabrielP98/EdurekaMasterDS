**1. Load the data from “college.csv” that has attributes collected about private and public colleges for a particular year. We will try to predict the private/public status of the college from other attributes**

```python
import numpy as np # linear algebra
import pandas as pd # data processing

#Visualisation frameworks
import matplotlib.pyplot as plt
%matplotlib inline  
import seaborn as sns

#Tools
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv('C:/Users/gabri/Desktop/Data Science/Python/Python for Data Science - Edureka/Module 9 - Supervised Learning 2/College.csv')
print(df.shape)
df.head()
```

Let's format our target variable:
* Yes = 1;
* No = 0.

```python
df.Private.unique()
```

```python
df['Private'] = df['Private'].replace('Yes', 1)
df['Private'] = df['Private'].replace('No', 0)
print(df['Private'].unique())
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

**2. Use LabelEncoder to encode the target variable in to numerical form and split the data such that 20% of the data is set aside fortesting.**

```python
#As we already enconded manually due low number of values, let's just split the data
X = df.drop('Private',axis=1)
y = df['Private']

X.head()
```

```python
# Split into train/test
X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.2, stratify=y,
                     random_state=0)

print('X_train shape:', X_train.shape, '\nX_test shape:', X_test.shape)
```

**3. Fit a linear svm from scikit learn and observe the accuracy. [Hint:Use Linear SVC]**

```python
from sklearn.svm import LinearSVC
```

```python
model = LinearSVC(random_state=0)
model.fit(X_train,y_train)
# Predict the value of the digit on the test subset
predicted = model.predict(X_test)

print(f"Classification report for classifier {model}:\n"
      f"{metrics.classification_report(y_test, predicted)}\n")
```

**4. Preprocess the data using StandardScalar and fit the same model again and observe the change in accuracy.**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
model2 = make_pipeline(StandardScaler(),
                         LinearSVC(random_state=0))

model2.fit(X_train, y_train)
```

```python
predicted2 = model2.predict(X_test)
```

```python
print(f"Classification report for classifier {model2}:\n"
      f"{metrics.classification_report(y_test, predicted2)}\n")
```

**5. Use scikit learn’s gridsearch to select the best hyperparameter for a non-linear SVM,identify the model with best score and its parameters.**

```python
from sklearn.svm import SVC
```

```python
model3 = make_pipeline(StandardScaler(), SVC())

model3.fit(X_train, y_train)

predicted3 = model3.predict(X_test)

print(f"Classification report for classifier {model3}:\n"
      f"{metrics.classification_report(y_test, predicted3)}\n")
```

Let's use a more sophisticated technique so it doesn't take too long:

```python
from skopt import forest_minimize
from sklearn.model_selection import cross_val_score
```

```python
def tune_xgbr(params):

    
    C = params[0] 
    gamma = params[1] 

    print(params, '\n')
    #Model to be optimized
    mdl = make_pipeline(StandardScaler(), SVC(C=C,gamma=gamma))

    #Cross-Validation in order to avoid overfitting
    mse = cross_val_score(mdl, X_train, y_train, cv = 5, scoring = 'neg_mean_squared_error')
    mse = (-mse)
    print(mse.mean())
    return mse.mean()
```

```python
# Creating a sample space in which the initial randomic search should be performed
param_grid = [(0.01, 1), # C
          (0.1, 5)] # gamma
              
# Minimization using a random forest with 10 random samples and 40  iterations for forest minimize optimization.
result = forest_minimize(tune_xgbr, param_grid, random_state=1, n_calls=50, n_initial_points=10, n_jobs=-1, verbose=3)
```

```python
model3 = make_pipeline(StandardScaler(), SVC(C=0.981, gamma=0.145))

model3.fit(X_train, y_train)

predicted3 = model3.predict(X_test)

print(f"Classification report for classifier {model3}:\n"
      f"{metrics.classification_report(y_test, predicted3)}\n")
```
